```python
#Question1

import pandas as pd
df=pd.read_csv('/content/Salary_Data (4).csv')
df
```

```python
#target
y=df['Salary']
print(y)
print(y.shape)
```

```python
#feature
x=df['YearsExperience']
print(x)
print(x.shape)
```

```python
#normalize feature
xx=(x-x.min())/(x.max()-x.min())
print(xx)
```

```python
#normalize target
yy=(y-y.min())/(y.min()-y.max())
print(yy)
```

```python
#graph
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.show()
```

```python
target_var=data2['class']
target_var.head()
```

```python

features=data2.drop('class',axis=1)
features.head()
```

```python
for column in features.columns:
                       plt.figure(figsize=(10,6))
plt.scatter(features[column], target_var, alpha=0.5)
plt.xlabel(column)
plt.ylabel('Target')
plt.title(f'Relationship between {column} and Target')
plt.show()
```

```python
one_values = ["Male", "Yes"]
zero_values = ["Female","No"]
for column in features.columns:
                       features[column] = features[column].replace(to_replace=one_values, value=1)
features[column] = features[column].replace(to_replace=zero_values, value=0)
features
```
